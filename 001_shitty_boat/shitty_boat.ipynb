# Shitty Boat Problem

We built a shitty boat that will sink and kill everyone onboard if the weight capacity exceeds 1000 kg.
100 children have bought tickets to ride the boat, but we cannot allow all of them onboard.
We will let them onboard in the order they bought their tickets, until their total weight exceeds the limit.
All of them are Americans, so they only know their weight in pounds.
Write a function that:
1. Prints the total weight of the passengers
2. creates a dictionary of the passengers in the format of:
    ** *Mr/Ms* *name*: *weight in kg* **


## The Data

You have three sets of data:

* A list of names
* A dictionary that relates names to a gender
* A dictionary that relates names to a weight

```python
# The list of names
from random import shuffle

names = list()

with open('male_names.txt') as mans, open('female_names.txt') as womans:
    names.extend([man[:-2] for man in mans])
    names.extend([woman[:-2] for woman in womans])

shuffle(names)
```

```python
# Names : Gender
gender = dict()
mens = dict()
womens = dict()

with open('male_names.txt') as mans, open('female_names.txt') as womans:
    mens = {man[:-2]: 'Male' for man in mans}
    womens = {woman[:-2]: 'Female' for woman in womans}
    
gender = {**mens, **womens}
```

```python
# Names : Weight in lb
from random import randrange

weights = {name: randrange(20,51) for name in names}
```

```python
# All the data
from pprint import pprint

print('Names')
print('-' * 30)
pprint(names)
print('\n')

print('Gender')
print('-' * 30)
pprint(gender)
print('\n')

print('Weights')
print('-' * 30)
pprint(weights)
print('\n')
```

```python
kiloweight = dict()
def kilograms(lbs):
    return lbs * 0.453592
newnames = dict()
totalweight = 0

for people in names:
    kilos = kilograms(weights[people])
    if totalweight + kilos <1000:
        if gender[people] == 'Male':
            totalweight += kilos
            kiloweight['Mr ' + people] = kilos
        else:
            totalweight += kilos
            kiloweight['Ms ' + people] = kilos    
    else:
        break
pprint(kiloweight)
print(totalweight)
    
```
